####Loops
Loops repeat a specified section of code until a stopping condition is reached. **While-loops** only specify the stopping condition directly. **For-loops** specifiy a loop-variable, a stopping condition, and a change to the loop-variable to be made every iteration. For and While loops can be used intercangeably. 
Examples:


```
a = [20,43,6,90,78,3]
print("forwards")
for i in range(len(a)):
  print(a[i])

print("now backwards:")
i = len(a)-1
while(i>=0):
  print(a[i])
  i -= 1
```

## 2D Data
We're going to plot time series data from the National Oceanic and Atmospheric Administration (NOAA). [The data](https://www.star.nesdis.noaa.gov/sod/lsa/SeaLevelRise/LSA_SLR_timeseries.php) shows the global change in mean sea level from the 1990s to the present, taken across 4 different sattelite measurements with overlapping times of operation. 

```
url = "https://www.star.nesdis.noaa.gov/sod/lsa/SeaLevelRise/slr/slr_sla_gbl_keep_txj1j2_90.csv"
df = pd.read_csv(url,skiprows=5)
time = df["year"].values

# data was collected using 4 satellite radar sensors
topex = df["TOPEX/Poseidon"].values
jason1 = df["Jason-1"].values
jason2 = df["Jason-2"].values
jason3 = df["Jason-3"].values

# finding the problem values in each sensor's data, NaN (Not a Numer)
nan0 = np.isnan(topex)
nan1 = np.isnan(jason1) 
nan2 = np.isnan(jason2) 
nan3 = np.isnan(jason3)

# making time vectors specific to each sensor, only at the values that are NOT NaN.
# only plotting values from each sensor that are NOT NaN.

plt.rcParams['figure.figsize'] = [10, 8] # sets the figure size
plt.plot(time[~nan0],topex[~nan0]);
plt.plot(time[~nan1],jason1[~nan1]);
plt.plot(time[~nan2],jason2[~nan2]);
plt.plot(time[~nan3],jason3[~nan3]);

plt.title("Global Rising Sea-Levels")
plt.xlabel("Year")
plt.ylabel("Change in mean sea level (mm)")
plt.legend(["TOPEX","Jason-1","Jason-2","Jason-3"]);
```

## Loading the data

The Boston housing data set was collected in the 1970s to study the relationship between house price and various factors such as the house size, crime rate, socio-economic status, etc.  Since the variables are easy to understand, the data set is ideal for learning basic concepts in machine learning.  The raw data and a complete description of the dataset can be found on the UCI website:

https://archive.ics.uci.edu/ml/datasets/Housing

First, use `pd.read_csv` command to read the data from the file located at

https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data

```
import pandas as pd
import numpy as np
names =[
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 
    'AGE',  'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'PRICE']

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data', 
                 header=None,delim_whitespace=True,names=names,na_values='?')

df.head(6)  # print the first six examples
```

## Target Vector

Create a response vector `y` with the values in the column `PRICE`.  The vector `y` should be a 1D `numpy.array` structure.

```
df.columns.tolist()
y = df['PRICE']
```

Use the response vector `y` to find the mean house price in thousands and the fraction of homes that are above $40k. (You may realize this is very cheap.  Prices have gone up a lot since the 1970s!).   

```
mean = np.mean(y)
j = 0;
for i in range(len(y)):
    if y[i] >= 40:
        j = j+1      
percent = (j/len(y))*100
print('The mean house price is', mean, 'thousands of dollars')
print('Only', percent, 'percent are above $40k')
```

## Visualizing the Data

Similar to the `y` vector, create a predictor vector `x` containing the values in the `RM` column, which represents the average number of rooms in each region.

```
x = df['RM']
```

Create a scatter plot of the price vs. the `RM` attribute. 

```
plt.plot(x,y,'o')
plt.xlabel('Average number of rooms in a region')
plt.ylabel('Price')
plt.grid(True)
```

We will fit a line to these data points. Roughly estimate the slope and y-intercept of the line that best fits this data.

Now, Replot the scatter plot above, but now with the line on top of the above plot

```
beta1 = 9
beta0 = -30
xplt = np.array([3,9]) 
yplt = beta1*xplt + beta0
plt.plot(x,y,'o')    # Plot the data points
bottom, top = plt.ylim()
plt.ylim((bottom, top))
plt.plot(xplt,yplt,'-',linewidth=3)  # Plot the regression line
plt.xlabel('Average number of rooms in a region')
plt.ylabel('Price')
plt.grid(True)
```

There are several metrics that can measure the goodness of the fit. We will discuss about them as we go along. 
The (normalized) residual sum of squares, $$NRSS=\frac{1}{N} \sum_{i=1}^{N} (y_i-\hat{y}_i)^2$$
   is one such metric shown below.

```
yhat=beta0+beta1*x
NRSS = np.mean((y-yhat)**2)
print("RSS = {0:7.2f}".format(NRSS))
```
